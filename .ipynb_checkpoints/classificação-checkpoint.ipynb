```python
def porcentage(column_input, dataframe):
    df_temp = dataframe.groupby([column_input, 'Survived']).size().reset_index(name='Count').merge(dataframe[column_input].value_counts().reset_index(), on=column_input).copy()
    df_temp['porcentagem'] = round((df_temp['Count']/df_temp['count'])*100,2)
    return df_temp
```

```python
porcentage("Sex",df)
```

```python
porcentage("Pclass",df)
```

```python
porcentage("Embarked",df)
```

```python
porcentage("Parch",df)
```
